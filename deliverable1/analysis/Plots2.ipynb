```python
from __future__ import print_function 
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib as mpl 
from datetime import datetime

mpl.rcParams['savefig.dpi'] = 100 
mpl.rcParams['figure.dpi'] = 100
mpl.rcParams.update({'font.size': 40})

%config InlineBackend.figure_format = 'retina' 
%matplotlib inline

import matplotlib.dates as mdates
```

```python
with open("results/dist_fare_amount") as f:
    lines = f.readlines()

tuples = [line.strip() for line in lines]

fare = [int(x[1:].split(",")[0]) for x in tuples]
rides = [int(x[:-1].split(",")[1]) for x in tuples]
pct_rides = [x / sum(rides) for x in rides]
```

```python
plt.hist(fare, weights=pct_rides, range=(0,60), bins=12)
plt.title('Fare Amount Frequency')
plt.xlabel('Fare (Dollars)')
plt.ylabel('Proportion of Rides')
```

```python
plt.hist(fare, weights=pct_rides, range=(60,200), bins = 28)
plt.title('Fare Amount Frequency')
plt.xlabel('Fare (Dollars)')
plt.ylabel('Total Rides')
```

```python
with open("results/monthly_revenues.txt") as f:
    lines = f.readlines()

tuples = [line.strip() for line in lines]

dates = [datetime.strptime(x[1:].split(",")[0].replace("'",''), '%Y/%m') for x in tuples]
count = [int(x.split(",")[1][2:]) for x in tuples]
revenue = [float(x.split(',')[2][0:-2]) for x in tuples]

avg_revenue = [x[0]/x[1] for x in zip(revenue, count)]
```

```python
plt.plot(dates, revenue)
plt.title('Monthly Revenue Over Time')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
```

```python
plt.plot(dates, avg_revenue)
plt.title('Average Revenue Per Ride')
plt.xlabel('Month')
plt.ylabel('Average Revenue')
```

```python
with open("../results/dist_trip_distance") as f:
    lines = f.readlines()

tuples = [line.strip() for line in lines]
```

```python
distance = [int(x.split(',')[0][1:]) for x in tuples]
count = [int(x.split(',')[1][1:-1]) for x in tuples]
pct_count = [x / sum(count) for x in count]
```

```python
plt.hist(distance, weights=pct_count, range=(0,20), bins=20)
plt.title('Trip Distance Distribution')
plt.xlabel('Miles')
plt.ylabel('Proportion of Trips')
```
